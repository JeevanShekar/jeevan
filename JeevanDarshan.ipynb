```python
#line chart
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 3, 4, 3, 5])
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#cosine_wave
x = np.arange(0, 20, 0.2)
y = np.cos(x)
plt.plot(x, y)
plt.title('Cosine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude = cos(time)')
plt.show()
```

```python
x = np.arange(0,2*np.pi, 2*np.pi/100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
